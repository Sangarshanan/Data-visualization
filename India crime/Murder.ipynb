```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
data = pd.read_csv('32_Murder_victim_age_sex.csv')
data.head()
```

```python
data['Sub_Group_Name'].unique()
```

```python
df = data[(data['Year'] == 2001)]
plt.figure(figsize=(10,6))
sns.barplot(df['Area_Name'], df['Victims_Total'], alpha=0.8)
plt.xticks(rotation='vertical')
plt.xlabel('States', fontsize=14)
plt.ylabel('Number of Murders', fontsize=14)
plt.title("Number of Murders with respect to each state", fontsize=16)
plt.show()
```

```python
a=[]
for i in range(2001,2011):
    yr = data[(data['Year'] == i)]
    a.append(sum(yr['Victims_Total']))
plt.figure(figsize=(10,6))
plt.plot(data['Year'].unique(),a, color = 'red')
plt.title('Number of Murders in India from 2001 to 2010')
plt.xlabel('Year')
plt.ylabel('Number of rapes')
plt.show()  
```
